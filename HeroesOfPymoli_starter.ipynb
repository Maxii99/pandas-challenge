```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count


* Display the total number of players


```python
purchase_df = pd.DataFrame(purchase_data)

# Calculte total players usig length and unique functions

total_players = len(purchase_df["SN"].unique())

# display total_players in datafram
total_df = pd.DataFrame({"Total Players": [total_players]})
total_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#  calculation to obtain total number of unique items
unique_items=len(purchase_data["Item ID"].unique())

# calculation for average price and format it to $0.00
average_price1 =round(purchase_data["Price"].mean(),2)

formatted_avp=format(average_price1, ",.2f")

average_price= "$"+ formatted_avp

# caculations for total purchases

total_purchases= len(purchase_data["Purchase ID"].unique())

# calculate total revenue by summing up all prices and format total purchase to $0.00

total_revenue1=round(purchase_data["Price"].sum(),2)

formatted_revenue = format(total_revenue1, ",.2f")
total_revenue = "$" + formatted_revenue

# display summary data in dataframe

summary_df=pd.DataFrame({"Number of Unique Items": unique_items, "Average Price": "$"+str(average_price1), 
                         "Number of Purchases": [total_purchases], "Total Revenue": (total_revenue)})
summary_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Age Demographics
age_data = purchase_data[["Age", "SN"]].drop_duplicates()

# # create bins
age_bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 999.9]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

age_data["Age Group"]=pd.cut(age_data["Age"], age_bins, labels=group_names)

#  Two main calculations
# 1. total count
age_total = age_data["Age Group"].value_counts()

#  2. percentage of players
age_percent = np.round(age_total/total_players*100, decimals =2)

# summary age demographics dataframe
age_summary_df = pd.DataFrame({"Total Count": age_total, "Percentage of Players": age_percent.map("{:.2f}%".format)})
age_summary_df.sort_index()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# bins for purchasing data
purchase_data["Age"] = pd.cut(purchase_data["Age"], age_bins, labels=group_names)

# total purchase count
age_purchase = purchase_data.groupby(["Age"]).count()["Purchase ID"]

# Average purchase price
age_average_price = purchase_data.groupby(["Age"]).mean()["Price"]

# Total purchase value by age
age_val_purchase = purchase_data.groupby(["Age"]).sum()["Price"]

# average total purchase
age_ave_total = age_val_purchase/age_total

#  summary gender purchase dataframe

age_summary_purchase_df = pd.DataFrame({"Purchase Count":age_purchase.map('{:.2f}'.format),
                           "Average Purchase Price": age_average_price.map('${:.2f}'.format),
                           "Total Purchase Value": age_val_purchase.map('${:.2f}'.format),
                           "Average Total Purchase Per Person": age_ave_total.map('${:.2f}'.format)})
age_summary_purchase_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Run basic calculations:

# Purchase count
spender_count = purchase_data.groupby(['SN']).count()['Purchase ID']

# Avg purchase price
spender_average = purchase_data.groupby(['SN']).mean()['Price']

# Total purchase value
spender_total = purchase_data.groupby(['SN']).sum()['Price']

# Create data frame using above values
spender_df = pd.DataFrame({'Purchase Count': spender_count, 'Average Purchase Price': spender_average, 'Total Purchase Value': spender_total})

# Sort by total purchase value
sorted_spender_df = spender_df.sort_values('Total Purchase Value', ascending=False)

# Format avg purchase price
sorted_spender_df['Average Purchase Price'] = sorted_spender_df['Average Purchase Price'].map('${:,.2f}'.format)

# Format total purchase value
sorted_spender_df['Total Purchase Value'] = sorted_spender_df['Total Purchase Value'].map('${:,.2f}'.format)

# Set values and return dataframe of top five spenders
sorted_spender_df = sorted_spender_df.loc[:,['Purchase Count', 'Average Purchase Price', 'Total Purchase Value']]
sorted_spender_df.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Retrieve the Item ID, Item Name, and Item Price columns
item_data = purchase_data.loc[:,['Item ID', 'Item Name', 'Price']]
item_data.head()
item_data_df = pd.DataFrame(item_data)

# Run basic calculations:

# Purchase count
item_count = item_data.groupby(['Item ID', 'Item Name']).count()['Price']

# Total purchase value
item_total_value = item_data.groupby(['Item ID', 'Item Name']).sum()['Price']

# Create data frame using above values
pop_df = pd.DataFrame({'Purchase Count': item_count, 'Total Purchase Value': item_total_value})

# Join the two data frames on Item ID
popular_items_df = item_data_df.merge(pop_df, how='left', on=['Item ID'])

# Sort by Purchase Count
popular_items_df.sort_values(['Purchase Count'], ascending=[False], inplace=True)

# Format Price and Total Purchase Value columns
popular_items_df['Price'] = popular_items_df['Price'].map('${:,.2f}'.format)
popular_items_df['Total Purchase Value'] = popular_items_df['Total Purchase Value'].map('${:,.2f}'.format)

# Drop duplicate rows based on Item ID and display the most popular products
popular_items_df.drop_duplicates(['Item ID']).head()
```
